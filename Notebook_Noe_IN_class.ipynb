#### Create a new column

We want to create a new column named **full_price** that takes into account the price of the listing + the cleaning fee

```python
paris_df["full_price"] = paris_df["price"] + paris_df.cleaning_fee
```

```python
#Check again this new column
paris_df.head()
```

#### Create a new column with a condition (.loc)

 Let's say as a PM you are interested in having a closer look to luxury listings. A better way to display apartment price is just to list an apartment as : Affordable Or Expensive. So let's create a column called : **price_category**
 
 If the price is more than 300 then the listing is "Expensive" otherwise it is affordable

```python
paris_df.loc[paris_df.price > 300,"price_category"] = "Expensive"
paris_df.loc[paris_df.price <= 300,"price_category"] = "Affordable"
```

```python
#check columns again
paris_df[paris_df.price_category =="Affordable"]
paris_df.price_category.value_counts()
```

#### What is the share of expensive listings for arr. with most listings ‚ùì

```python
# Let's set the size right away
plt.figure(figsize=(15, 8))

# countplot with a secondary dimension 
sns.countplot(data=paris_df,x="zipcode",hue="price_category")
```

We got the same problem ... but now the way to fix it is by using the **order** parameter of the countplot method

```python
my_order = paris_df.zipcode.value_counts().head(10).index
sns.countplot(data=paris_df,x="zipcode",hue="price_category",order=my_order)
```

#### Violin Plot

#### What is the price distribution for each arrondissement ‚ùì

```python
#set the size
plt.figure(figsize=(15, 8))

#Set an order
order = paris_df.zipcode.value_counts().head(10).index

#plot
sns.violinplot(data=affordable, x="zipcode", y="price", order=order)
```
