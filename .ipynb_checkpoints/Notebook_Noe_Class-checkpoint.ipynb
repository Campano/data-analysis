#### Create a new column

We want to create a new column named **full_price** that takes into account the price of the listing + the cleaning fee

```python
paris_df["full_price"] = paris_df["price"] + paris_df["cleaning_fee"]
```

#### Create a new column with a condition (.loc)

 Let's say as a PM you are interested in having a closer look to luxury listings. A better way to display apartment price is just to list an apartment as : Affordable Or Expensive. So let's create a column called : **price_category**
 
 If the price is more than 300 then the listing is "Expensive" otherwise it is affordable

```python
paris_df.loc[paris_df["price"] > 300, "price_category"] = 'Expensive'
paris_df.loc[paris_df["price"] <= 300, "price_category"] = 'Affordable'
```

```python
paris_df.columns
```

###  3Ô∏è‚É£ Clean the Data

##### Drop/Fill the empty values

```python
#fill the missing host_response_rate with "0%" 
paris_df["host_response_rate"].fillna("0%",inplace=True)
```

```python
#drop the listings with no description
paris_df.dropna(subset = ["description"], inplace=True)
```

```python
#checking again the size to see if it has got smaller
paris_df.shape
```

##### Keep only relevant columns

```python
#check the columns
paris_df.columns
```

```python
#drop summary, description and listing_url
columns_to_drop = ["summary","description","listing_url"]
paris_df.drop(columns_to_drop,axis="columns", inplace=True)
paris_df.columns
```

##### Change value types 

```python
#change response rate to float
paris_df.host_response_rate = paris_df.host_response_rate.str.replace('%', '').astype(float)

#check if changes were indeed made
paris_df.head(5)
```

### Visualize Data

#### What are arrondissement with most listings ‚ùì

```python
plt.figure(figsize=(15, 8))
#Let's get the dataframe that we want to plot 
arr = paris_df.zipcode.value_counts()

#Let's plot it 
arr.plot(kind='bar')
```

That's not very nice... We have too many bars ! 

```python
#Select only 10 arrondissement with most listings
arr = paris_df.zipcode.value_counts().head(10)
arr.plot(kind='bar')
```

That's still not so nice ! Let's make it bigger ! 

```python
plt.figure(figsize=(15, 8))
arr.plot(kind='bar')
```

#### What is the share of expensive listings for arr. with most listings ‚ùì

```python
# Let's set the size right away
plt.figure(figsize=(15, 8))

# countplot with a secondary dimension 
sns.countplot(data=paris_df,x="zipcode",hue="price_category")
```

We got the same problem ... but now the way to fix it is by using the **order** parameter of the countplot method

```python
my_order = paris_df.zipcode.value_counts().head(10).index
sns.countplot(data=paris_df,x="zipcode",hue="price_category",order=my_order)
```

#### [Distribution]¬†What is the distribution of listings' price  ‚ùì

```python
#Set the size
plt.figure(figsize=(15, 8))

#plot
sns.distplot(affordable['price'])
```

#### Violin Plot

#### What is the price distribution for each arrondissement ‚ùì

```python
#set the size
plt.figure(figsize=(15, 8))

#Set an order
order = paris_df.zipcode.value_counts().head(10).index

#plot
sns.violinplot(data=affordable, x="zipcode", y="price", order=order)
```
